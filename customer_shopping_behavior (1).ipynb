```python
df['Review Rating'] = df['Review Rating'].fillna(df['Review Rating'].mean())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 30, 40, 60, 120],       # age boundaries
    labels=['Young-Adult', 'Adult', 'Middle-Aged', 'Senior']   # category names
)
```

```python
print(df[['age', 'age_group']].head(10))
```

```python
#Convert purchase frequency into estimated number of days between purchases
freq_map = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Annually': 182,
    'Annually': 365,
    'Every 3 Month':90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(freq_map)
```
